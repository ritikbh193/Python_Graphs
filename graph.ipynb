### Subplots

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


# Creating subplots (3 rows, 2 columns)
fig, axs = plt.subplots(3, 2, figsize=(14, 15))

# Histogram for total_sqft
sns.histplot(df1['price'], kde=True, ax=axs[0, 0], color='skyblue')
axs[0, 0].set_title('Distribution of Prices')

# Bar plot for the number of bathrooms
sns.countplot(x='bath', data=df1, ax=axs[0, 1], palette='viridis')
axs[0, 1].set_title('Number of Bathrooms')
axs[0, 1].set_xlabel('Bathrooms')

# Scatter plot for total_sqft vs price
sns.scatterplot(x='total_sqft', y='price', data=df1, ax=axs[1, 0], hue='area_type', palette='coolwarm', s=100)
axs[1, 0].set_title('Total Sqft vs Price')
axs[1, 0].set_xlabel('Total Sqft')
axs[1, 0].set_ylabel('Price (in lakhs)')

# Bar plot for balcony counts
sns.countplot(x='balcony', data=df1, ax=axs[1, 1], palette='magma')
axs[1, 1].set_title('Number of Balconies')
axs[1, 1].set_xlabel('Balcony')

# Pie chart for area_type distribution
area_counts = df1['area_type'].value_counts()
axs[2, 0].pie(area_counts, labels=area_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
axs[2, 0].set_title('Distribution of Area Types')

# Line plot for price trend with respect to total_sqft
df_sorted = df1.sort_values('total_sqft')
axs[2, 1].plot(df_sorted['total_sqft'], df_sorted['price'], marker='o', color='purple', linestyle='-')
axs[2, 1].set_title('Price Trend with Total Sqft')
axs[2, 1].set_xlabel('Total Sqft')
axs[2, 1].set_ylabel('Price (in lakhs)')

# Adjust layout
plt.tight_layout()
plt.show()
```
